# **Matplotlib**

```python
# Esperanza de vida de cada pais
life_exp = gapminder['life_exp']

gdp_cap = gapminder['gdp_cap']

pop = gapminder['population']
```

```python
gdp_cap.head()
```

```python
# Grafica normal
# plt.style.use('_mpl-gallery-nogrid')
plt.plot(gdp_cap, life_exp)
plt.show()
```

```python
# Grafica de dispersion
#plt.scatter(gdp_cap, life_exp)

plt.scatter(gdp_cap, life_exp, alpha = 0.4, color = '#28F575')
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([50, 80, 100], ['50', '80', '100'])

plt.show()
```

```python
plt.scatter(pop, life_exp, alpha = 0.8)
plt.show()
```

```python
plt.hist(life_exp, bins=20, alpha = 0.8, color='#28F575')
plt.show()
plt.clf()
```
