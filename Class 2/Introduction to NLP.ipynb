```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('all')
```

```python
example = "Hoy es un dia increible, ya que vamos a aprender a procesar texto con\
 Python. Es importante conocer nuevos temas de python, y las librerias mas\
 utilizadas en NLP para python para procesar texto"
```

Tokenizar el texto es dividirlo en las unidades que lo conforman.

```python
example.split()
```

```python
words = word_tokenize(example)
print(words)
```

Normalizar. Convertir cada elemento de la lista en minusculas y eliminar los simbolos contenidos de nuestro interes

```python
words_lower = [ token.lower() for token in words if token.isalpha() ]
print(words_lower)
```

Esta sentencia de codigo hace lo mismo que la celda anterior

```python
words_lower = []
for token in words:
  if token.isalpha():
    words_lower.append(token.lower())

print(words_lower)
```

Vamos a eliminar las palabras vacias

```python
filtered_sentences = [ w for w in words_lower if w not in stopwords.words('spanish') ]
print(filtered_sentences)
```

```python
stop_words = set(stopwords.words('spanish'))

filtered_sentences = []
for i in words_lower:
  if i not in stop_words:
    filtered_sentences.append(i)

print(filtered_sentences)
```

```python
from collections import Counter
```

```python
Counter(filtered_sentences).most_common(4)
```

Palabra raiz

```python
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
porter_stemmer = PorterStemmer()
snowball_stemmer = SnowballStemmer('spanish')
```

```python
example2 = ['felicidad', 'pensando', 'caminando', 'escuchando',
           'reina', 'cantante']
```

```python
for i in example2:
  print(snowball_stemmer.stem(i))
```

```python
example3 = ['happines', 'thinking', 'python', 'tuning', 'happy',
            'food', 'dangerous', 'queen']
```

```python
for i in example3:
  print(porter_stemmer.stem(i))
```

Etiquetas gramaticas o tipos semanticos

```python
example4 = set(words_lower)

tag = nltk.pos_tag(example4)
print(tag)
```
